```python
patients=df.groupby('PatientId')['RecordNum'].apply(list).reset_index(name="Record_list")
patients_dict=patients.set_index('PatientId').to_dict()
patients_dict=patients_dict['Record_list']
print(patients_dict)
```

```python
import psycopg2
import urllib3
import certifi
from bs4 import BeautifulSoup
import re
import os
import requests
import csv
import requests

import warnings
warnings.filterwarnings('ignore', message='Unverified HTTPS request')

path='./FinalRecordsSeg/'
baseurl = 'https://physionet.org/physiobank/database/mimic3wdb/matched/'
proxies = {
  'http': 'http://cis-india-pitc-bangalore.corporate.ge.com:80',
  'https': 'https://cis-india-pitc-bangalore.corporate.ge.com:80',
}
patientIdList = list(set(df.PatientId.values))
N = len(patientIdList)
print('Number of patients : ' + str(N))
for i in range(N):
    PatientId = patientIdList[i]
    PatientIdS = 'p'+str(PatientId).zfill(6)
    link = baseurl + PatientIdS[:3] + '/' + PatientIdS + '/'
    #print(link)
    response = requests.get(link, verify=False,proxies=proxies)
    html= BeautifulSoup(response.content.decode('utf-8'))
    for rec in html.find_all('a'):
        record= rec.get('href')
        #print(record)
        pattern= re.compile(".*n.hea")
        pattern1= re.compile(".*n.dat")
        if re.match(pattern,record):
            #print(re.search(r'.*n.hea', record).group())
            match= re.search(r'.*n.hea', record).group()
            link_new= link+ match
            #print(link_new)
            r = requests.get(link_new, verify=False,proxies=proxies)
            soup = BeautifulSoup(r.content, 'html5lib')
            body=soup.body.string
            sub_id= re.search(r'(.*n).dat', body).group(1)
            
            if(int(sub_id[:-1]) in patients_dict[PatientId]):
                filename= sub_id + '.hea'
                body1= re.sub(r'^.*n', sub_id, body)
                body1= body1.encode()
                fl= path+filename
                if(not os.path.isfile(fl)):
                    print('Writing .hea File for patient SNo. '+str(i))
                    with open(fl, "wb") as file:
                        file.write(body1)
                #file.write(r.content)

        if re.match(pattern1,record):
            #print(re.search(r'.*n.dat', record).group())
            match= re.search(r'.*n.dat', record).group(0)
            if int(match[:-5]) in patients_dict[PatientId]:
                link_new= link+ match
                #print(link_new)
                r = requests.get(link_new, verify=False,proxies=proxies)
                fl= path+match
                if(not os.path.isfile(fl)):
                    print('Writing .dat File for patient SNo. '+str(i))
                    with open(fl, "wb") as file:
                        file.write(r.content)
```

```python
# Create a dataframe structure to add annotation for segmentation
for serial_id in range(np.shape(df)[0]):
    print(serial_id)
    [RRraw,SpO2raw,HRraw,RRlowess,SpO2lowess,HRlowess,Label] = getAllCols(df,serial_id,'./FinalRecordsSeg/')
    SegAnnoDf = pd.DataFrame(data={'RRraw':RRraw,'SpO2raw':SpO2raw,'HRraw':HRraw,'RRlowess':RRlowess,'SpO2lowess':SpO2lowess,'HRlowess':HRlowess,'Label':Label})
    myfn = str(serial_id)
    myfn = myfn.zfill(3)
    myfname = './SeriesSegData/p'+myfn+'.csv'
    SegAnnoDf.to_csv(myfname,index=False)
```

```python
RRrawlist = []
SpO2rawlist = []
HRrawlist = []
for serial_id in range(np.shape(df)[0]):
    #print(serial_id)
    myfn = str(serial_id)
    myfn = myfn.zfill(3)
    myfname = r'./SeriesSegData/p'+myfn+'.csv'
    SegAnnoDf = pd.read_csv(myfname, encoding='iso-8859-1')
    RRrawlist.extend(list(SegAnnoDf['RRraw'].values))
    SpO2rawlist.extend(list(SegAnnoDf['SpO2raw'].values))
    HRrawlist.extend(list(SegAnnoDf['HRraw'].values))
print(len(RRrawlist))
print(len(SpO2rawlist))
print(len(HRrawlist))
```

```python
RRmean = np.mean(RRrawlist)
SpO2mean = np.mean(SpO2rawlist)
HRmean = np.mean(HRrawlist)
print(RRmean)
print(SpO2mean)
print(HRmean)
RRstd = np.std(RRrawlist)
SpO2std = np.std(SpO2rawlist)
HRstd = np.std(HRrawlist)
print(RRstd)
print(SpO2std)
print(HRstd)
```
